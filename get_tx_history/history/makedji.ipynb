```python
jsonFilePath = r'mini_dow.json'


def readjson(jsonFilePath, w_path):
    jarr = []
    new_jarr = []

    with open(jsonFilePath, encoding='utf-8') as j:
        jarr = json.load(j)
        # print(arr[1]['time'][11:16], j, type(j), type(arr[0]))
    dayObj = {}
    for i in jarr:
      # print('open', i['time'][11:16])
      if(i['time'][11:16]=='18:00'):
        # print('open', i['time'][11:16])
        dayObj['start'] = i
      elif(i['time'][11:16]=='16:30'):
        # print('close', i['time'][11:16])
        dayObj['end'] = i
        dayObj['percent'] = (float(dayObj['end']['close']) - float(dayObj['start']['open'])) / float(dayObj['start']['open']) * 100
        if(i['time'][20:] == '05:00'):
          dayObj['is_summer'] = False
        else:
          dayObj['is_summer'] = True

        print(dayObj)
        new_jarr.append(dayObj)
        dayObj = {}
    with open(w_path, 'w', encoding='utf-8') as jsonf: 
      jsonString = json.dumps(new_jarr, indent=4)
      jsonf.write(jsonString)
    
jsonFilePath = r'mini_dow.json'
w_path = r'dow_open_close_percent.json'
readjson(jsonFilePath, w_path)
```
